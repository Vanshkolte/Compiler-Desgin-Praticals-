```python
class TACGenerator:
    def __init__(self):
        self.temp_count = 0
        self.label_count = 0
        self.tac = []

    def new_temp(self):
        self.temp_count += 1
        return f't{self.temp_count}'

    def new_label(self):
        self.label_count += 1
        return f'L{self.label_count}'

    def translate_expression(self, expression):
        if '=' in expression:
            left, right = expression.split('=')
            left = left.strip()
            right = right.strip()

            if '+' in right:
                operands = right.split('+')
                operand1 = operands[0].strip()
                operand2 = operands[1].strip()
                temp_var = self.new_temp()
                self.tac.append(f"{temp_var} = {operand1} + {operand2}")
                self.tac.append(f"{left} = {temp_var}")
            else:
                self.tac.append(f"{left} = {right}")

    def generate_tac(self, condition, then_branch, else_branch):
        condition_temp = self.new_temp()
        self.tac.append(f"{condition_temp} = {condition}")

        then_label = self.new_label()
        else_label = self.new_label()
        end_label = self.new_label()

        self.tac.append(f"if {condition_temp} goto {then_label}")
        self.tac.append(f"goto {else_label}")

        self.tac.append(f"{then_label}:")
        for statement in then_branch:
            self.translate_expression(statement)

        self.tac.append(f"goto {end_label}")

        self.tac.append(f"{else_label}:")
        for statement in else_branch:
            self.translate_expression(statement)

        self.tac.append(f"{end_label}: END")

    def display_tac(self):
        for code in self.tac:
            print(code)


def main():
    tac_generator = TACGenerator()
    condition = input("Enter the condition (e.g., 'a < 5'): ")

    print("Enter the then statements (type 'done' when finished):")
    then_branch = []
    while True:
        stmt = input("> ")
        if stmt.lower() == 'done':
            break
        then_branch.append(stmt)

    print("Enter the else statements (type 'done' when finished):")
    else_branch = []
    while True:
        stmt = input("> ")
        if stmt.lower() == 'done':
            break
        else_branch.append(stmt)

    tac_generator.generate_tac(condition, then_branch, else_branch)
    tac_generator.display_tac()


if __name__ == "__main__":
    main()
```
