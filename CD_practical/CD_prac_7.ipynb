```python
import re

class TACOptimizer:
    def __init__(self, tac):
        self.tac = tac  
        self.optimized_tac = []  
        self.variable_values = {}  
        self.variable_copy = {}  

   
    def copy_propagation(self, line):
        for var in self.variable_copy:
            line = re.sub(rf'\b{var}\b', self.variable_copy[var], line)
        return line

    def constant_propagation(self, line):
        for var, value in self.variable_values.items():
            line = re.sub(rf'\b{var}\b', str(value), line)
        return line

   
    def constant_folding(self, line):
        match = re.search(r'(\d+)\s*([\+\-\*/])\s*(\d+)', line)
        if match:
           
            operand1, operator, operand2 = int(match.group(1)), match.group(2), int(match.group(3))
            result = eval(f'{operand1} {operator} {operand2}')
  
            line = line.replace(match.group(0), str(result))
        return line

   
    def common_subexpression_elimination(self):
        seen_expressions = {}
        for i, line in enumerate(self.optimized_tac):
            match = re.search(r'(t\d+)\s*=\s*(.+)', line)
            if match:
                temp_var, expression = match.group(1), match.group(2)
                if expression in seen_expressions:
                
                    original_var = seen_expressions[expression]
                    self.optimized_tac[i] = f'{temp_var} = {original_var}'
                else:
                   
                    seen_expressions[expression] = temp_var


    def apply_optimizations(self):
        changes = True
        while changes:
            changes = False
            self.optimized_tac = [] 
            for line in self.tac:
                original_line = line

                line = self.copy_propagation(line)

       
                line = self.constant_propagation(line)

                line = self.constant_folding(line)

               
                copy_match = re.match(r'(\w+)\s*=\s*(\w+)', line)
                if copy_match:
                    var, value = copy_match.groups()
                    self.variable_copy[var] = value

                constant_match = re.match(r'(\w+)\s*=\s*(\d+)', line)
                if constant_match:
                    var, value = constant_match.groups()
                    self.variable_values[var] = int(value)

              
                if line != original_line:
                    changes = True

        
                self.optimized_tac.append(line)

          
            self.common_subexpression_elimination()

       
            self.tac = self.optimized_tac

    def get_optimized_tac(self):
        return self.optimized_tac




if __name__ == "__main__":

    tac = [
        "t1 = 4 + 5",
        "t2 = t1",
        "t3 = t2 + 3",
        "t4 = t1",
        "t5 = 4 + 5"
    ]

    optimizer = TACOptimizer(tac)
    optimizer.apply_optimizations()
    optimized_tac = optimizer.get_optimized_tac()

    print("Optimized TAC:")
    for line in optimized_tac:
        print(line)
```
